```python
from pydantic import ConfigDict, BaseModel
```

```python

class Date(BaseModel):
    year: int
    month: int
    day: int

class Config(BaseModel):
    model_config = ConfigDict(strict=True)
    latitude: float
    longitude: float
    radius: int
    limit: int
    sensor_name: str
    from_date: Date
    to_date: Date
```

```python
config = Config.model_validate(config)
```
